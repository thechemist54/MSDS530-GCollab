```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset from seaborn
iris = sns.load_dataset("iris")
```

```python
# Display the first few rows of the dataset
print(iris.head())
```

```python
# Scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Box plots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
sns.boxplot(data=iris, x="species", y="sepal_length", ax=axes[0, 0])
sns.boxplot(data=iris, x="species", y="sepal_width", ax=axes[0, 1])
sns.boxplot(data=iris, x="species", y="petal_length", ax=axes[1, 0])
sns.boxplot(data=iris, x="species", y="petal_width", ax=axes[1, 1])
plt.suptitle("Box Plots of Iris Features by Species")
plt.tight_layout()
plt.show()
```

```python
# Histograms
iris.hist(by="species", figsize=(10, 8))
plt.suptitle("Histograms of Iris Features by Species")
plt.tight_layout()
plt.show()
```

```python
# Violin plots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
sns.violinplot(data=iris, x="species", y="sepal_length", ax=axes[0, 0])
sns.violinplot(data=iris, x="species", y="sepal_width", ax=axes[0, 1])
sns.violinplot(data=iris, x="species", y="petal_length", ax=axes[1, 0])
sns.violinplot(data=iris, x="species", y="petal_width", ax=axes[1, 1])
plt.suptitle("Violin Plots of Iris Features by Species")
plt.tight_layout()
plt.show()
```

```python
# Bar plot
sns.countplot(data=iris, x="species")
plt.title("Distribution of Iris Species")
plt.show()
```

```python
# Line plot
sns.lineplot(data=iris, x=iris.index, y="sepal_length")
plt.title("Trend of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
# Pair plot
sns.pairplot(data=iris, hue="species")
plt.title("Pairwise Relationships of Iris Features")
plt.show()
```

```python
# Pie chart
species_counts = iris["species"].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Proportion of Iris Species")
plt.axis("equal")
plt.show()
```
