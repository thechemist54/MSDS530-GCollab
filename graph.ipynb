```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a sample dataset of text documents
dataset = [
    "Python is an interpreted high-level general-purpose programming language.",
    "Java is a class-based object-oriented programming language that is designed to have as few implementation dependencies as possible.",
    "C++ is a general-purpose programming language created by Bjarne Stroustrup as an extension of the C programming language.",
    "JavaScript is a scripting or programming language that allows you to implement complex features on web pages.",
    "SQL is a domain-specific language used in programming and designed for managing data held in a relational database management system."
]
```

```python
# Extract keywords from the documents using NLP techniques (simplified for this example)
keywords = [
    ["Python", "interpreted", "high-level", "general-purpose", "programming", "language"],
    ["Java", "class-based", "object-oriented", "programming", "language", "implementation", "dependencies"],
    ["C++", "general-purpose", "programming", "language", "Bjarne Stroustrup", "extension", "C"],
    ["JavaScript", "scripting", "programming", "language", "web pages"],
    ["SQL", "domain-specific", "language", "programming", "data", "relational database"]
]
```

```python
# Create a graph using networkx
G = nx.Graph()
```

```python
# Add nodes for each keyword
for doc in keywords:
    for word in doc:
        G.add_node(word)
```

```python
# Add edges for co-occurrence of keywords in the same document
for doc in keywords:
    for i in range(len(doc)):
        for j in range(i+1, len(doc)):
            G.add_edge(doc[i], doc[j])
```

```python
# Draw the graph using matplotlib
plt.figure(figsize=(10, 10))
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Create a graph using an adjacency list
graph = {
  'A' : ['B','C'],
  'B' : ['D', 'E'],
  'C' : ['F'],
  'D' : [],
  'E' : ['F'],
  'F' : []
}
```

```python
 #Convert the graph to a networkx object
G = nx.DiGraph(graph)

# Perform depth first search and store the traversal order
order = list(nx.dfs_preorder_nodes(G, source='A'))

# Draw the graph with different colors for visited and unvisited nodes
colors = ['red' if node in order else 'blue' for node in G.nodes()]
nx.draw(G, with_labels=True, node_color=colors)
plt.show()
```
